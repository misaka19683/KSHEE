首先，我需要计算这个模型在电场为0和存在(1,1,1)方向电场的情况
我的目前是验证前面这两种情况下，此模型是luttinger liquid
验证方法是计算三个量
1. 计算能量密度，使能量密度满足 $E(r)=E_U(r)+(-)^r E_A(r)$, 其中的 $E_U(r)$ 是一致项的能量密度，$E_A(r)$ 是交替项的能量密度。
    其中的 $E_A(r)$ 在长程极限下，有 $E_A(r)\propto (r_L)^{-\kappa}$， 还有其中的 $\kappa$ 是Luttinger参数，并且有 $r_L=\frac{L}{\pi}\sin(\frac{\pi r}{L})$，在OBC条件下计算，96格点

2. 计算central charge 也是一种测试 Luttinger liquid 的测试，我们可以从纠缠熵中提取出central charge, 而根据conformal field theory(CFT)的理论，有式子$S_L(x)=\lambda \frac {c}{ 3} \ln \left [ \frac{L}{\pi}\sin(\frac{\pi x}{L}) \right ]$，而目的是计算 c ，同时在OBC和PBC条件下计算，看看$c=1$是否成立
我们的模型具有一个两格点的周期性，因此我们只写出在第一个unit cell的哈密顿量

The Hamiltonian for the spin-1/2 $KJ\Gamma\Gamma^\prime K_2 J_2$ chain is given by
$$
\begin{aligned}
1. \quad & KS_1^x S_2^x + J(S_1^x S_2^x+S_1^y S_2^y+S_1^z S_2^z)+\Gamma(S_1^y S_2^z+S_1^z S_2^y)+\Gamma^\prime(S_1^x S_2^y+S_1^y S_2^x+S_1^x S_2^z+S_1^z S_2^x)+K_2 S_1^z S_3^z+J_2 (S_1^x S_3^x+S_1^y S_3^y+S_1^z S_3^z) \\
2. \quad & KS_2^y S_3^y + J(S_2^x S_3^x+S_2^y S_3^y+S_2^z S_3^z)+\Gamma(S_2^z S_3^x+S_2^x S_3^z)+\Gamma^\prime(S_2^y S_3^z+S_2^z S_3^y+S_2^x S_3^y+S_2^y S_3^x)+K_2 S_2^z S_4^z+J_2 (S_2^x S_4^x+S_2^y S_4^y+S_2^z S_4^z)
\end{aligned}
$$

现在我们需要添加电场影响的项
对于 x 方向的电场
$$
\begin{aligned}
1. \quad & E_x\left [ \lambda_D^{(out)}(S_1^z S_2^x - S_1^x S_2^z + S_1^x S_2^y - S_1^y S_2^x )+\lambda_\Gamma^{(out)} (S_1^x S_2^y +S_1^y S_2^x+S_1^x S_2^z+S_1^z S_2^x) \right ] \\
2. \quad & E_x (\lambda_{D+}^{(in)}-\lambda_{D-}^{(in)}) (S_2^z S_3^x - S_2^x S_3^z)
\end{aligned}
$$
对于 y 方向的电场，哈密顿量有
$$
\begin{aligned}
1. \quad & E_y (\lambda_{D+}^{(in)}+\lambda_{D-}^{(in)}) (S_1^y S_2^z-S_1^z S_2^y) \\
2. \quad & E_y \lambda_D^{(out)}(S_2^x S_3^y - S_2^y S_3^x + S_2^y S_3^z - S_2^z S_3^y)+ E_y \lambda_\Gamma^{(out)} (S_2^y S_3^z + S_2^z S_3^y + S_2^y S_3^x + S_2^x S_3^y)
\end{aligned}
$$
对于 z 方向电场，哈密顿量有
$$
\begin{aligned}
1. \quad E_z (\lambda_{D+}^{(in)}-\lambda_{D-}^{(in)}) (S_1^y S_2^z-S_1^z S_2^y)\\
2. \quad E_z (\lambda_{D+}^{(in)}+\lambda_{D-}^{(in)}) (S_2^z S_3^x-S_2^x S_3^z)
\end{aligned}
$$

$$

dimerization 序参量的定义如下：

II Dimerization

six-sublattice rotation is define as
$$
\begin{align}
Sublattice 1: (x,y,z) &\to (-x^\prime,-y^\prime,z^\prime) \\
Sublattice 2: (x,y,z) &\to (x^\prime,z^\prime,-y^\prime) \\
Sublattice 3: (x,y,z) &\to (-y^\prime,-z^\prime,x^\prime) \\
Sublattice 4: (x,y,z) &\to (y^\prime,x^\prime,-z^\prime) \\
Sublattice 5: (x,y,z) &\to (-z^\prime,-x^\prime,y^\prime) \\
Sublattice 6: (x,y,z) &\to (z^\prime,y^\prime,-x^\prime)
\end{align}
$$

坐标反转之后得到的是
$$
\begin{align}
1. &(x^\prime,y^\prime,z^\prime) \to (-x,-y,z) \\
2. &(x^\prime,y^\prime,z^\prime) \to (x,-z,y) \\
3. &(x^\prime,y^\prime,z^\prime) \to (z,-x,-y) \\
4. &(x^\prime,y^\prime,z^\prime) \to (y,x,-z) \\
5. &(x^\prime,y^\prime,z^\prime) \to (-y,z,-x) \\
6. &(x^\prime,y^\prime,z^\prime) \to (-z,y,x)
\end{align}
$$
in six-sublattice rotated frame ,dimerization order parameter $O_i^{\prime dim}$ 被定义为
$$
O_i^{\prime dim} = \vec S_i^{\prime} \cdot \vec S_{i+1}^{\prime} -\vec S_{i+1}^\prime \cdot \vec S_{i+2}^\prime
$$
Its form $O_i^{dim}$ in the unrotated frame is
$$
\begin{align}
O_{1+2n}^{dim} &= (-S_{1+2n}^y S_{2+2n}^y + S_{1+2n}^z S_{2+2n}^x + S_{1+2n}^x S_{2+2n}^z) -(-S_{1+2n}^x S_{2+2n}^x + S_{1+2n}^y S_{2+2n}^z + S_{1+2n}^z S_{2+2n}^y) \\
O_{2+2n}^{dim} &= (-S_{1+2n}^x S_{2+2n}^x + S_{1+2n}^y S_{2+2n}^z + S_{1+2n}^z S_{2+2n}^y) -(-S_{1+2n}^y S_{2+2n}^y + S_{1+2n}^z S_{2+2n}^x + S_{1+2n}^x S_{2+2n}^z)
\end{align}
$$

```python
# Cell 3: 定义参数 (这是你主要交互的地方)
import os
import pandas as pd

theta = np.pi * 0.44
phi = np.pi * 0.85

# 创建结果保存目录
os.makedirs("results/data", exist_ok=True)
os.makedirs("results/figures", exist_ok=True)

def save_task_data(results, task_name):
    """保存计算结果。OBC 保存 kappa 和 central charge，PBC 保存 central charge 和 dimerization。"""
    summary_path = f"results/data/{task_name}_summary.csv"
    bc_type = results.get('bc_type', 'open')
    
    data = {
        'task': [task_name],
        'central_charge': [results['cc_res']['c_avg']]
    }
    
    if bc_type == 'open':
        data['kappa'] = [results.get('kappa')]
    else: # periodic
        if 'dimerization' in results:
            data['dimer_avg_abs'] = [np.mean(np.abs(results['dimerization']))]
        
    pd.DataFrame(data).to_csv(summary_path, index=False)
    
    # 保存详细的键能量数据 (对于分析基础很有用)
    energy_path = f"results/data/{task_name}_energy.csv"
    pd.DataFrame({'E_bonds': results['E_bonds']}).to_csv(energy_path, index=False)
    
    # 保存详细的 dimerization 数据 (仅限 PBC)
    if bc_type == 'periodic' and 'dimerization' in results:
        dimer_path = f"results/data/{task_name}_dimer.csv"
        pd.DataFrame({'dimerization': results['dimerization']}).to_csv(dimer_path, index=False)
        
    print(f"Results for {task_name} ({bc_type}) saved.")

# 基础参数配置
base_params = {
    'L': 96,
    'bc_MPS': 'finite',
    'lattice_params': {'bc': 'open'},
    'gamma': np.sin(theta) * np.sin(phi), 'k': np.sin(theta) * np.cos(phi), 'j': np.cos(theta),
    'gamma_prime': 0.1, 'k2': 0.02, 'j2': 0.2,
    'lambda_in_dp': -0.1, 'lambda_in_dn': -0.05,
    'lambda_out': 0.04, 'lambda_gamma_out': 0.05,
}

dmrg_params = {
    'mixer': True,
    'max_E_err': 1.e-10,
    'trunc_params': {'chi_max': 1000, 'svd_min': 1.e-10},
}
```
